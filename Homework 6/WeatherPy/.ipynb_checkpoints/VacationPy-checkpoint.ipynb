```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_df = pd.read_csv("./output_data/city_weather_data.csv")
# Dropped extra index column
weather_df.drop(axis=1,labels='Unnamed: 0',inplace=True)
weather_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Cofigure map with api key and set locations from data frame columns
gmaps.configure(api_key=g_key)
heat_locations = weather_df[["Lat", "Lng"]]
```

```python
# Some formating to try get the map to center
figure_layout = {'600px': '500px', '1': '0 auto 0 auto'}
fig = gmaps.figure(center=(0,0), zoom_level=1.6, layout=figure_layout)
# Add heat layer using humidity as weights and 100% as max humidity
heat_layer = gmaps.heatmap_layer(heat_locations, weights=weather_df['Humidity'], 
                                 max_intensity=100,
                                 dissipating = False,
                                 point_radius = 2.5)

fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Setting idea conditions between 60-80 F and low winds low humidity
hotel_df = weather_df.loc[(weather_df["Max Temp"]>60)&(weather_df["Max Temp"]<80)&
                          (weather_df["Wind Speed"]<15)&(weather_df["Humidity"]<30)]
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Add empty column named hotel name
hotel_df["Hotel Name"]=''
hotel_df
```

```python
# Iterate through each row of hotel_df, for each row search for the closing lodging lcation and store into hotel name
for index, row in hotel_df.iterrows():
    # Setting query parameters
    target_coordinates = f"{row['Lat']}, {row['Lng']}"
    target_radius = 5000
    target_type = "lodging"
    params = {
        "location": target_coordinates,
        "radius": target_radius,
        "type" :target_type,
        "key": g_key
    }  
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(base_url, params=params).json()
    try:
        # If json traversal succeeds then store info
        hotel_df.at[index,"Hotel Name"]=response['results'][0]['name']
    except:
        # If json traversal fails then no hotels were found and drop current row
        print("no lodging were found")
        hotel_df.drop(index,inplace=True)
```

```python
# Setting varible reference to use the code from next cell
narrowed_city_df = hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in narrowed_city_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
# Setting a new figure
fig2 = gmaps.figure(center=(0,0), zoom_level=1.6, layout=figure_layout)
# Create a symbol layer, using parts of code from class activity and the two variable from the cell above
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    scale=5,
    info_box_content=hotel_info
)
# Add layers to fig2
fig2.add_layer(hotel_layer)
# Add heat layer created earlier
fig2.add_layer(heat_layer)
# Display Map
fig2
```
