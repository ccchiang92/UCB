```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
drug_df = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
combine_df = pd.merge(mouse_df, drug_df, on="Mouse ID")

# Display the data table for preview
combine_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_mean = combine_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
# # Convert to DataFrame
tumor_df = drug_time_mean.to_frame()
tumor_df.reset_index(inplace=True)
# # Preview DataFrame
tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
drug_time_sem = combine_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
drug_time_sem_df = drug_time_sem.to_frame()
drug_time_sem_df.reset_index(inplace=True)
# Preview DataFrame
drug_time_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
tumor_df_new_index = tumor_df.set_index("Timepoint")
tumor_group = tumor_df_new_index.groupby("Drug")
tumor_df_new=pd.DataFrame()
for group in tumor_group.groups:
    tumor_df_new[group]=tumor_group.get_group(group)["Tumor Volume (mm3)"]
tumor_df_new.head()
```

```python
# Generate the Plot (with Error Bars)
#Reformat sem df
tumor_sem_group = drug_time_sem_df.set_index("Timepoint").groupby("Drug")
tumor_sem_new=pd.DataFrame()
for group in tumor_sem_group.groups:
    tumor_sem_new[group]=tumor_sem_group.get_group(group)["Tumor Volume (mm3)"]

fig1, ax1 = plt.subplots()
dot_list=["o"]
label_list=[]
i=0
for drug in tumor_df_new.columns:
    ax1.errorbar(tumor_df_new.index, tumor_df_new[drug], tumor_sem_new[drug], fmt="o",label=drug)
    

# Save the Figure
"""
Search handle, legend, dot list
"""
```
